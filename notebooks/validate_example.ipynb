# End-to-end validation with `run_agent`

This notebook walks through the full `agentic_validation` workflow:

1. Build a `TaskInput` describing what you want to validate.
2. Call `run_agent` to run the structured reasoning, critique, SMT/Lean checking, and repair loop.
3. Inspect the returned `AgentResult`.

Install the package first if you have not already:

```bash
pip install -e .
```

````python
from agentic_validation import TaskInput, run_agent
````

## 1. Define the task

`TaskInput` captures everything the agent needs.  Only `task_id` and `goal` are required; all other fields fall back to package defaults (see `src/agentic_validation/schemas.py`).

````python
task = TaskInput(
    task_id="demo-arithmetic",
    goal="Validate that x > 5 implies x + 1 > 6",
    require_symbolic_checking=True,
)
print(task.model_dump_json(indent=2))
````

## 2. Run the agent

````python
result = run_agent(task)
````

## 3. Inspect the result

````python
print("Status:", result.verification_status)
print("Answer:", result.final_answer)
````

````python
import json
print(json.dumps(result.model_dump(), indent=2))
````

## Key fields in `AgentResult`

| Field | Description |
|---|---|
| `verification_status` | Final gate decision (`verified`, `rejected`, `unknown`) |
| `final_answer` | Human-readable conclusion |
| `accepted_steps` | Reasoning steps that passed critique |
| `failed_steps` | Steps that were rejected or could not be repaired |
| `checker_artifacts` | Raw SMT/Lean checker output |
| `repair_history` | Repair attempts made during the loop |
| `summary_state` | Aggregated summary of the full run |
